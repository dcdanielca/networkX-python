# Small Famous Graphs

```python
import networkx as nx
%matplotlib inline

import matplotlib.pyplot as plt
```

```python
p = nx.petersen_graph()

nx.draw(p, node_color='c', node_size=650, with_labels=True)
```

```python
t = nx.tutte_graph()
nx.draw(t, node_color='#A0CBE2', node_size=600, with_labels=300)
```

```python
s = nx.sedgewick_maze_graph()
nx.draw(s, node_color='#A0CBE2', node_size=700, with_labels=300)
```

```python
tt = nx.tetrahedral_graph()
nx.draw(tt, node_color='#A0CBE2', node_size=700, with_labels=300)
```

# Famous Social Networks Graphs

```python
k = nx.karate_club_graph()
nx.draw(k, node_color='#A0CBE2', node_size=650, with_labels=True)
```

```python
print(k.is_directed())
```

```python
print(nx.info(k))
```

```python
plt.figure(figsize=(10,7))
dsw = nx.davis_southern_women_graph()
nx.draw(k, node_color='c', node_size=700, with_labels=True)
```

```python
print(nx.info(dsw))
```

```python
print(dsw.edges(data = True))
```

```python
f = nx.florentine_families_graph()
nx.draw(f, node_color='g', node_size=900, with_labels=True)
```

```python
print(nx.info(f))
```

```python
print(f.edges(data=True))
```

```python
print(f.nodes(data=True))
```

# Classical Graphs

```python
barbel = nx.barbell_graph(10, 10)
nx.draw(barbel, node_color='b', node_size=800, with_labels=True)
```

```python
print(nx.info(barbel))
```

```python
loli = nx.lollipop_graph(10, 20)
nx.draw(loli, node_color='c', node_size=600, with_labels=True)
```

```python
print(nx.info(loli))
```

```python
binom = nx.binomial_graph(100, 0.15)
nx.draw(binom, node_color='r', node_size=700, with_labels=True)
```

```python
print(nx.info(binom))
```

```python
chordal = nx.chordal_cycle_graph(20)
nx.draw(chordal, node_color='b', node_size=500, with_labels=True)
```
