```python
import networkx as nx
%matplotlib inline

import matplotlib.pyplot as plt
```

```python
# adding attributes to one graph
```

```python
cities = nx.Graph()
```

```python
cities.add_edge('San Diego', 'L.A', distance= 0.4)
cities.add_edge('New York', 'Nashville', distance = 5.6)
cities.add_edge('Boston', 'D.C', distance = 0.8)
```

```python
cities.edges(data = True)
```

```python
nx.draw_networkx(cities, node_color='#9999FF')
```

```python
# creating a direct graph
```

```python
b = nx.DiGraph([('a','b'),('b', 'c'),('c', 'a')])
```

```python
# adding edges to the di-graph
b.add_edge('a','b')
b.add_edges_from([('d','c'),('d','b')])
```

```python
# plot graph
nx.draw_networkx(b, node_size=800)
```

```python
# print the out-degrees of the di-graph
print(b.out_degree())
```

```python
# print the in-degrees of the di-graph
print(b.in_degree())
```

```python
# print some other properties of the graph
print(b.in_edges())
print(b.out_edges())
```

# More on DiGraphs

```python
dg = nx.DiGraph()
```

```python
dg.add_edges_from([(1,2), (1,3),(2,4),(2,5),(2,6),(3,7),(7,8)])
```

```python
nx.draw_networkx(dg, node_size=750)
```

```python
print(nx.degree_centrality(dg))
```

```python
print(nx.betweenness_centrality(dg))
```

```python
print(nx.closeness_centrality(dg))
```
