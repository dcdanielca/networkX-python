# Multi-Graphs

```python
import networkx as nx
%matplotlib inline

import matplotlib.pyplot as plt
```

```python
G = nx.MultiGraph()
```

```python
G.add_node(1)
G.add_nodes_from([2,3])
G.add_nodes_from(range(100,110))
```

```python
nx.draw_networkx(G, node_color='c' , node_size=900)
```

```python
GM = nx.MultiGraph(day='Friday', month='December', year=2015)
```

```python
GM.graph
```

```python
GM.add_node(1, time='10 PM')
```

```python
GM.node[1]
```

```python
GM.add_edge(1 , 2, weigth = 4.7)
```

```python
GM.add_edges_from([(3,4),(4,5)], color='red')
```

```python
nx.draw_networkx(GM, node_color='green', node_size=900)
```

```python
GM.nodes(data = True)
```

```python
GM.edges(data = True)
```

# Multi DiGraphs

```python
G = nx.MultiDiGraph()
```

```python
G.add_node(1)
G.add_nodes_from([2,3])
G.add_nodes_from(range(100,110))
nx.draw(G, node_color='#A0CBE2', node_size=800, with_labels= True)
```

```python
GM = nx.MultiDiGraph(day = 'Sunday', month = 'January', year = 2015, DOB = '02/14/1985')
```

```python
GM.graph
```

```python
GM.add_node(1, time = '11:03 PM')
GM.node[1]
```

```python
GM.add_edge(1, 2, weigth=100)
GM.add_edges_from([(3,4),(4,5)], color = 'green')
```

```python
nx.draw_networkx(GM, node_color='c', node_size=200)
```
