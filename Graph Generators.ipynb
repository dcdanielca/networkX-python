# Graph Generators

```python
import networkx as nx
%matplotlib inline

import matplotlib.pyplot as plt
```

```python
# draw a complete graph
plt.figure(figsize=(10,7))
cg = nx.complete_graph(5, create_using = nx.DiGraph())
nx.draw_networkx(cg)
```

```python
print(cg.in_degree())
```

```python
print(cg.out_degree())
```

```python
# draw a graph called path-graph
plt.figure(figsize=(10,7))
pg = nx.path_graph(5, create_using=nx.DiGraph())
nx.draw_networkx(pg)
```

```python
# draw a gnr-graph, which is like facebook or twitter
plt.figure(figsize=(10,7))
gnr = nx.gnr_graph(100, 1, create_using=nx.DiGraph())
nx.draw_networkx(gnr)
```

```python
print(gnr.in_degree())
```

```python
# draw complete-bipartite-graph
plt.figure(figsize=(10,7))
cbg = nx.complete_bipartite_graph(20, 15)
nx.draw_networkx(cbg)
```

```python
print(cbg.order())
```

```python
print(cbg.size())
```

```python
# draw grid_graph
gg = nx.grid_graph(dim=[5,5])
nx.draw_networkx(gg, node_color='cyan', node_size=800)
```

```python
print(gg.order())
```

```python
print(gg.size())
```
