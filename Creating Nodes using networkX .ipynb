```python
import networkx as nx
%matplotlib inline

import matplotlib.pyplot as plt
```

```python
# create a an empty graph with no nodes and no edges

G = nx.Graph()
```

```python
# add a single node

G.add_node(1)
```

```python
# add a list of nodes

G.add_nodes_from([2,3])
```

```python
nx.draw_networkx(G, node_color = 'green', node_size= 700)
```

```python
# adding edges to a graph
G.add_edge(1, 2)
```

```python
nx.draw_networkx(G)
```

```python
# add edges using unpack edge tuples

e = {2 , 3}
G.add_edge(*e)
```

```python
nx.draw_networkx(G)
```

```python
# adding list of edges 

G.add_edges_from([{1,2},{1,3}])
```

```python
nx.draw_networkx(G)
```

```python
# find number of nodes in a graph
print(G.number_of_nodes())
```

```python
# find number of edges in a graph
print(G.number_of_edges())
```

```python
# get some properties of a graph
```

```python
# getting nodes in a graph
print(G.nodes())
```

```python
# getting edges in a graph
print(G.edges())
```

```python
# getting the neighbor of a connected graph
print(list(G.neighbors(1)))
```

```python
# get the degrees from one graph or one node
print(G.degree())
print(G.degree(1))
```

```python
# find if there exists one edge between two nodes
print(G.has_edge(2, 3))
print(G.has_edge(2, 4))
```

```python
# find if there exist a node in a graph
print(G.has_node(1))
print(G.has_node(5))
```

```python
# node manipulation
```

```python
H = nx.Graph()
```

```python
edgelist = [{0,1}, {1,2},{2,3},{0,3}]
```

```python
H.add_edges_from(edgelist)
```

```python
nx.draw_networkx(H, with_labels=True, node_color='cyan', node_size=700)
```

```python
# Remove edge
H.remove_node(3)
nx.draw_networkx(H, with_labels=True, node_color='cyan', node_size=700)
```

```python
# remove a list of nodes
H.remove_nodes_from({0, 1})
nx.draw_networkx(H, with_labels=True, node_color='cyan', node_size=700)
```
