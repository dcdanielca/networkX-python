# Social Network Analysis

```python
import networkx as nx
from datetime import datetime
import matplotlib.pyplot as plt
import numpy as np


%matplotlib inline
```

```python
er = nx.erdos_renyi_graph(100,0.15)
nx.draw(er, with_labels=True)
```

```python
sub = nx.subgraph(er, {54,35,45,76,59,90,4,99,29})
nx.draw(sub, node_color='c', node_size=800, with_labels=True)
```

# Find the number of triangles in a graph 

```python
G = nx.complete_graph(5)

nx.draw(G, with_labels=True)
```

```python
# print number of triangles at node 0
print(nx.triangles(G,0))
```

```python
# print number of triangles in the graph
print(nx.triangles(G))
```

```python
# print the list of triangle in the graph
print(list(nx.triangles(G,(0,1)).values()))
```
