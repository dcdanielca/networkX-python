```python
import networkx as nx
%matplotlib inline

import matplotlib.pyplot as plt
```

# Basic drawing using python functions

```python
# define a function in python
def draw_graph():
    G = nx.Graph()
    G.add_edges_from([(1,2), (2,3), (1,3), (1,4)])
    nx.draw(G, node_size=800, node_color='cyan', with_labels=True)
    
```

```python
# call the draw_graph function
draw_graph()
```

```python
# define a new function call draw_barasi_albert
def draw_barabasi_albert():
    G = nx.barabasi_albert_graph(100, 10)
    hst = nx.degree_histogram(G)
    plt.hist(hst, bins=20, color='red')
```

```python
draw_barabasi_albert()
```

# Let us say a Graph G = {(A,B), (B,C), (C,A)} , where each tuple in the set denotes an edge

How to create and visualize this graph G

```python
# define function called simple-graph

def simple_graph():
    #create empty graph
    G = nx.Graph()
    
    #add edges
    G.add_edge('A', 'B')
    G.add_edge('B', 'C')
    G.add_edge('C', 'A')
    
    #draw the graph
    nx.draw(G, with_labels=True, node_color='y', node_size=800)
```

```python
# call function
simple_graph()
```

# How to delete nodes from a graph

```python
def delete_nodes_in_graph():
    G = nx.Graph()
    
        
    #add edges
    G.add_edge('A', 'B')
    G.add_edge('B', 'C')
    G.add_edge('C', 'A')
    
    #draw the graph
    nx.draw(G, with_labels=True, node_color='y', node_size=800)
    
    # remove node B from the graph
    G.remove_node('B')
    
    nx.draw(G, with_labels=True, node_color='c', node_size=800)
    
```

```python
delete_nodes_in_graph()
```

# Delete specific edges {(A,B), (A,C)} from G

```python
def detele_specific_edges_in_graph():
    G = nx.Graph()
    
    #add edges
    G.add_edge('A', 'B')
    G.add_edge('B', 'C')
    G.add_edge('C', 'A')
    
    nx.draw(G, with_labels=True)
    plt.show()
    
    # remove 2 edges from the graph
    G.remove_edge('A','B')
    G.remove_edge('A','C')
    
    nx.draw(G, with_labels=True)
    plt.show()
```

```python
detele_specific_edges_in_graph()
```

# Customize node size and node colors in the graph

```python
def custom_node_size_and_node_color():
    G = nx.Graph()
    
    #add edges
    G.add_edge('A', 'B')
    G.add_edge('B', 'C')
    G.add_edge('C', 'A')
    
    nx.draw(G, with_labels=True)
    plt.show()

    # define the nodelist
    nodelst = ['A', 'B', 'C']
    
    # define the colorList
    colorlst = ['c','m','y']
    
    # define the nodesizeList
    nodesize = [1000, 2000, 3000]
    
    nx.draw(G, with_labels=True, node_color=colorlst, node_size=nodesize, nodelist=nodelst)
    plt.show()
```

```python
custom_node_size_and_node_color()
```
