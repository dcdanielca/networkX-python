# Draw a random graph called grid_2d_graph

```python
import networkx as nx
%matplotlib inline

import matplotlib.pyplot as plt
```

```python
G = nx.grid_2d_graph(4,4) #4x4 grid

pos = nx.spring_layout(G, iterations=100)

# first plot
plt.subplot(221)
nx.draw(G, pos, font_size=8)

# second plot
plt.subplot(222)
nx.draw(G, pos, node_color='k', node_size=0, with_labels=False)

# third plot 
plt.subplot(223)
nx.draw(G, pos, node_color='g', node_size=1, with_labels=False)

# fourth plot
plt.subplot(224)

#convert undirected graph to directed
H = G.to_directed()
nx.draw(H, pos, node_color='b', node_size=20)

plt.savefig('grids.png')

plt.show()
```

# Bipartite

Bipartite graphs are graphs that have two (bi-) partitions (-partite) of nodes. Nodes within each partition are not allowed to be connected to one another, rather, they can only be connected to nodes in other partition.

Bipartite graphs can be useful for modeling relations between two sets of entitities. 

```python
from networkx.algorithms import bipartite
```

```python
B = nx.Graph()
```

```python
# define cities
cities = ['Beijing', 'Xian', 'Vancouver', 'San Francisco', 'Austin', 'Boston']

# define people
people = ['Theo', 'Student']

# adding edges of cities and people
B.add_nodes_from(cities, bipartite='cities')
B.add_nodes_from(people, bipartite='people')

#define edges 
edges = [('Theo', 'Vancouver'), ('Student', 'Xian'), ('Theo', 'San Francisco'), ('Student', 'Boston'), ('Theo', 'Boston'),('Student','Beijing')]
                                                                                 
B.add_edges_from(edges)                                                                                 
nx.draw(B, with_labels=True, node_color='c', node_size=750)
```

# Drawing a house

```python
plt.figure(figsize=(12,10))
house = nx.house_graph()

# set positions of the nodes
pos = {0:(0,0), 1:(1,0), 2: (0,1), 3:(1,1), 4: (0.5,2)}

nx.draw_networkx_nodes(house, pos, node_size=2000, nodelist=[4])
nx.draw_networkx_nodes(house, pos, node_size=3000, nodelist=[0,1,2,3], node_color='c')
nx.draw_networkx_edges(house, pos, alpha=0.5, width=6)
```
